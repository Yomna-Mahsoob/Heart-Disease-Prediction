```python
import numpy as np

x_scaled = np.load('data/X_scaled.npy')
y = np.load('data/y.npy')
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
pca= PCA()
x_pca= pca.fit_transform(x_scaled)
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(explained_variance), marker='o', color='#409fc1')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Number of Components')
plt.grid(True)
plt.show()
```

```python
from matplotlib.colors import ListedColormap
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(x_scaled)

custom_cmap = ListedColormap(["#409fc1", "#818c94"])

plt.figure(figsize=(8,6))
plt.scatter(X_2d[:,0], X_2d[:,1], c=y, cmap=custom_cmap, alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('2D PCA Scatter Plot (Heart Disease Classes)')
plt.colorbar(label='Target (0 = No Disease, 1 = Disease)')
plt.grid(True)  
plt.show()
```
